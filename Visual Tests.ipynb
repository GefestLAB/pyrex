```python
# Standard imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# PyREx imports
from pyrex.digsig import AskaryanSignal
```

```python
# Signal plotting helper function
def plot_sig(signal, zoom=None, title=None, figsize=(6,4)):
    plt.figure(figsize=figsize)
    if zoom is not None:
        print("Not implemented")
    plt.plot(signal.times*1e9, signal.values)
    if title is not None: plt.title(title)
    plt.xlabel("Time (ns)")
```

# Test Askaryan Pulses

First make sure that the longitudinal shower profile makes sense

```python
dummy = AskaryanSignal([0], energy=1e5, theta=45*np.pi/180)
z_max = 2*dummy.max_length()
n_z = 100
z_vals, dz = np.linspace(0, z_max, n_z, endpoint=False, retstep=True)
Q = np.zeros(n_z)
for i, z in enumerate(z_vals):
    Q[i] = dummy.charge_profile(z)
plt.semilogy(z_vals, Q)
```

```python
times = np.linspace(0,100e-9,1001)
ask = AskaryanSignal(times, energy=1e5, theta=45*np.pi/180, t0=25e-9)
plot_sig(ask)
```
